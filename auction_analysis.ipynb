```python
dff[['sum_bid', 'max_bid','avg_bid']] = dff[['sum_bid', 'max_bid','avg_bid']].fillna( dff[['sum_bid', 'max_bid','avg_bid']].mean())
```

```python
dff['sum_bid'].unique()
```

```python
dff['win_rate_per_bid'] = dff['win_count']/dff['response_count'] *100
# dff['win_rate'] = dff['win_count']/dff['sum_bid']
# dff['win_rate']
```

```python
# dff_high = dff.loc[ ]
# dff['win_rate'].max()

dff_sorted = dff.sort_values(by='win_rate_per_bid', ascending=False)
```
